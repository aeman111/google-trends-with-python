## COMPARING PROGRAMMING LANGUAGES 

```python

# build list of keywords
kw_list = ["python", "ruby", "JavaScript", "java"] 

# build the payload
pytrend.build_payload(kw_list, cat=None, timeframe='2007-01-01 2021-03-01', geo='GB') 

# get the interest over time
df = pytrend.interest_over_time()

# plot prog lang trends over time
plt.figure(figsize=(20, 10))
plt.plot(df.index,df.python,'k*')
plt.plot(df.index,df.ruby,'r*')
plt.plot(df.index,df.JavaScript,'b*')
plt.plot(df.index,df.java,'g*')
plt.legend(["Python", "Ruby", "JavaScript", "Java"])
```
